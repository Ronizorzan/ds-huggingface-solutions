```python
from transformers import pipeline
from pandas import DataFrame
import json
```

```python
# Carregar o modelo pré-treinado para análise de sentimento
sentiment_classifier = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')
```

```python
sentiment_classifier("I love this movie! It's fantastic.")
```

```python
sentiment_classifier("I hate this movie. It's terrible.")
```

```python
sentiment_classifier("This movie is okay, not great but not bad either.")
```

```python
with open("../logs/execution_log.json", "r") as log_file:
    lines = log_file.readlines()
    logs = [json.loads(line) for line in lines]
df_execution_times = DataFrame(data=logs)
df_execution_times.rename(columns={"review_count": "Quantidade de reviews", "execution_time": "Tempo de execução (s)"}, inplace=True)
df_execution_times.sort_values(by="Quantidade de reviews", inplace=True)
df_execution_times.reset_index(drop=True, inplace=True)
```
